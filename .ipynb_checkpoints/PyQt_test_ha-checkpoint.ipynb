```python
import pyaudio
import scipy
import struct
import scipy.fftpack

# Testing SVC with live audio data

from hearing_assistive import *
from sklearn.externals import joblib
import numpy as np

%matplotlib inline

clf = joblib.load('SVC/svc_v2.pk1')


from Tkinter import *
import threading
import time, datetime
#import wckgraph
import math
from hearing_assistive import *

#ADJUST THIS TO CHANGE SPEED/SIZE OF FFT
bufferSize=132300
#bufferSize=2**8

# ADJUST THIS TO CHANGE SPEED/SIZE OF FFT
sampleRate=44100 
#sampleRate=64000
global chunks
p = pyaudio.PyAudio()
chunks=[]
ffts=[]
def stream():
        time.sleep(0.5)
        global chunks, inStream, bufferSize
        while True:
                chunks.append(inStream.read(bufferSize))
                

def record():
        global w, inStream, p, bufferSize
        inStream = p.open(format=pyaudio.paInt16,channels=1,\
                rate=sampleRate,input=True,frames_per_buffer=bufferSize)
        threading.Thread(target=stream).start()


def read_wave_str(z_str):
    sampwidth = 2
    dtype_map = {1:np.int8, 2:np.int16, 3:'special', 4:np.int32}
    if sampwidth not in dtype_map:
        raise ValueError('sampwidth %d unknown' % sampwidth)
    
    if sampwidth == 3:
        xs = np.fromstring(z_str, dtype=np.int8).astype(np.int32)
        ys = (xs[2::3] * 256 + xs[1::3]) * 256 + xs[0::3]
    else:
        ys = np.fromstring(z_str, dtype=dtype_map[sampwidth])

    # if it's in stereo, just pull out the first channel
    framerate = 44100

    #ts = np.arange(len(ys)) / framerate
    wave = Wave(ys, framerate=framerate)
    wave.normalize()

    return wave

def run_svm(var_wav):
    var_wav_l = [var_wav]
    spectrogram = var_wav.make_spectrogram(seg_length = 1764)
    features = spectrogram.mfcc()
    # convert to numpy array
    
    testset = np.asarray(features)

    # restructure testset before being processed by svm
    testset = testset.reshape(1,-1)
    print testset.shape
    var_pred = clf.predict(testset)
    return var_pred
    

def go(ui):
        global w,fftx,ffty,btn_bool
        print "STARTING!"
        threading.Thread(target=record).start()
        while btn_bool:
                if len(chunks)>0:
                    pya_var = chunks.pop(0)
                    pya_wav = read_wave_str(pya_var)
                    var_pred = run_svm(pya_wav)
                    if var_pred[0] == 1:
                        ui.label.setPixmap(ui.label.var_pix_AlarmFound)
                    elif var_pred[0] == 0:
                        ui.label.setPixmap(ui.label.var_pix_AlarmNotFound)

#for function go
#     passing in ui may not be best way to update gui
#     other possible solution use global var
#           or rewrite this as a function in app setup; may be best
```
