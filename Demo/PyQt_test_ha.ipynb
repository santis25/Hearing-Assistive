```python
import sys
from PyQt4 import QtGui , QtCore

class Window(QtGui.QMainWindow):
    def __init__(self):    #template
        super(Window, self).__init__()
        self.setGeometry(50,50,500,300)
        self.setWindowTitle("Hearing Assist")
        # main layout
        self.mainLayout = QtGui.QVBoxLayout()
        self.home()
        
    def home(self):              #view
        rec_pr = Rec_Prompt()
        self.mainLayout.addWidget(self, rec_pr)
        self.show()
        
    def rec_run(self):              #view
        btn = QtGui.QPushButton("Quit", self)
        btn.clicked.connect(self.close_application)
        btn.resize(btn.sizeHint())
        btn.move(100,100)
        self.raise_()
        
        
    def onFinished(self):
        self.pushButton.setText("done")
    def close_application(self):   #method
        print("whoooooaaa so custom!!")
        sys.exit()
        
class Rec_Prompt(QtCore.QObject):
    def __init__(self):    #template
        super(Rec_Prompt, self).__init__()
        self.init_p()
        
    def init_p(self):
        btn = QtGui.QPushButton("Start", self)
        btn.clicked.connect(self.run_recorder)
        btn.resize(btn.sizeHint())
        btn.move(100,100)
        
    def run_recorder(self):   #method
        command  = "./Sounds/record.py"
        args = ["./Sounds/rec_samp.wav" , "5"]
        process = QtCore.QProcess(self)
        process.finished.connect(self.onFinished)
        process.startDetached(command, args)
        
        
def run():
    app = QtGui.QApplication(sys.argv)
    GUI = Window()
    sys.exit(app.exec_())

run()


# def window():
#     app = QtGui.QApplication(sys.argv)
#     w = QtGui.QWidget()
#     b = QtGui.QLabel(w)
#     b.setText("Hello World!")
#     w.setGeometry(100,100,200,50)
#     b.move(50,20)
#     w.setWindowTitle("PyQt")
#     w.show()
#     sys.exit(app.exec_())
    
# if __name__ == '__main__':
#     window()
```
