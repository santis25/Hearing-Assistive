```python
import hear_assist_tools as ha

%matplotlib inline
```

```python
wave = ha.read_wave('Sounds/school_fire_alarm.wav')
wave.normalize()
wave.plot()
```

```python
wave.make_audio()
```

```python
start = 0.1
duration = 0.8
segment = wave.segment(start, duration)
segment.plot()
```

```python
segment.make_audio()
```

```python
spectrum = segment.make_spectrum()
spectrum.plot()
```

```python
spectrum.low_pass(2500)    # cutoff anything higher than frequency 2.5 kHz
spectrum.plot(high=5000)   # provide an upper bound at 5 kHz
```

```python
segment.normalize()
segment.apodize()
segment.plot()
```

```python
filtered = spectrum.make_wave()  # convert spectrum back to wave
filtered.normalize()
filtered.apodize()
filtered.plot()
```

```python
segment.make_audio()
```

```python
filtered.make_audio()
```
