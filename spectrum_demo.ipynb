```python
from hearing_assistive import *

import numpy as np
%matplotlib inline
```

```python
fs = 44100                                 # sampling rate
t = np.arange(0.0,0.002,1.0/fs)            # all points between -0.002 and 0.002 by step 1/fs
f0 = 1000                                  # frequency
phi = np.pi/2                              # phase
A = 1.0                                    # amplitude
signal = A*np.sin(2*np.pi*f0*t+phi)
```

```python
wave = Wave(signal, t, fs)
wave.plot()

spectrum = wave.make_spectrum()
spectrum.plot()
```

```python
print max(wave.ys)
print max(spectrum.amps)
```

```python
wave = read_wave('Sounds/school_fire_alarm.wav')
wave.normalize()
wave.plot()
```

```python
wave.make_audio()
```

```python
start = 0.1
duration = 0.8
segment = wave.segment(start, duration)
segment.plot()
```

```python
segment.make_audio()
```

```python
spectrum = segment.make_spectrum()
spectrum.plot()
```

```python
spectrum.low_pass(2500)    # cutoff anything higher than frequency 2.5 kHz
spectrum.plot(high=3000)   # provide an upper bound at 5 kHz
```

```python
segment.normalize()
segment.apodize()
segment.plot()
```

```python
filtered = spectrum.make_wave()  # convert spectrum back to wave
filtered.normalize()
filtered.apodize()
filtered.plot()
```
